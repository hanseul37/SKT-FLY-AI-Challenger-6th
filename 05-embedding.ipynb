```python
import os

os.environ['AZURE_OPENAI_API_KEY'] = ''
os.environ['OPENAI_API_VERSION'] = '2023-05-15'
os.environ['OPENAI_API_TYPE'] = 'azure'
os.environ['AZURE_OPENAI_ENDPOINT'] = 'https://labuser24-aiservice-002.openai.azure.com'
```

```python
from langchain.embeddings import AzureOpenAIEmbeddings

embedding_model = AzureOpenAIEmbeddings(
    azure_deployment='dev-text-embedding-ada-002',
    chunk_size = 500,
)
```

```python
embeddings = embedding_model.embed_documents(
    {
        '안녕하세요',
        '제 이름은 홍길동입니다.',
        '이름이 무엇인가요?',
        '랭체인은 유용합니다.',
        'Hello World'
        }
)
```

```python
len(embeddings[0])
```

```python
query_q = embedding_model.embed_query('이 대화에서 언급된 이름은 무엇인가요?')
query_a = embedding_model.embed_query('이 대화에서 언급된 이름은 홍길동 입니다.')
```

```python
import numpy as np

def cos_sim(A, B):
    return np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))
```

```python
print(cos_sim(query_q, query_a))
print(cos_sim(query_q, embeddings[1]))
print(cos_sim(query_q, embeddings[3]))
```

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings

model_name = 'BAAI/bge-small-en'
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}
hf = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
embeddings_sample = ([
    'today is monday',
    'weather is nice today',
    'whats the problem',
    'langchain is useful',
    'Hello world',
    'My name is morris'
])
```

```python
embeddings = hf.embed_documents(embeddings_sample)
```

```python
len(embeddings[0])
```

```python
query_q = hf.embed_query('Hello? who is this?')
query_a = hf.embed_query('Hi this is harrison.')
```

```python
print(cos_sim(query_q, query_a))
print(cos_sim(query_q, embeddings[1]))
print(cos_sim(query_q, embeddings[5]))
```

```python
model_name = 'jhgan/ko-sbert-nli'
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}

ko = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
embeddings_sample = ([
    '안녕하세요',
    '제 이름은 홍길동 입니다.',
    '이름이 무엇입니까?',
    '랭체인은 유용합니다.',
    '홍길동 아버지의 이름은 홍상직 입니다.'
])
```

```python
embeddings = ko.embed_documents(embeddings_sample)
```

```python
len(embeddings[0])
```

```python
qeury_q = ko.embed_query('홍길동은 아버지를 아버지라 무르지 못하였습니다. 홍길동 아버지의 이름은 무엇입니까?')
query_a = ko.embed_query('홍길동의 아버지는 엄했습니다.')
```

```python
print(cos_sim(qeury_q, query_a))
print(cos_sim(qeury_q, embeddings[1]))
print(cos_sim(qeury_q, embeddings[3]))
print(cos_sim(qeury_q, embeddings[4]))
```
