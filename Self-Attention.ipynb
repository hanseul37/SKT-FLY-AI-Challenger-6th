```python
import numpy as np

# 1) 입력 임베딩 X (3 x 4)
X = np.array([
    [0.2, 0.1, 0.5, 0.3],  # I
    [0.4, 0.3, 0.8, 0.1],  # am
    [0.7, 0.6, 0.1, 0.9]   # Sam
])

# 2) Q, K, V를 모두 X로 가정 (W_Q, W_K, W_V = I)
Q = X
K = X
V = X

# 3) QK^T
dot = Q.dot(K.T)
# 4) 스케일링 (d_k=4 -> sqrt(4)=2)
dk = 4
scaled_dot = dot / np.sqrt(dk)

# 5) 행별 softmax
def softmax(row):
    ex = np.exp(row)
    return ex / np.sum(ex)

A = np.apply_along_axis(softmax, 1, scaled_dot)

# 6) 어텐션 결과: A x V
Z = A.dot(V)

print("QK^T =\n", dot)
print("\nScaled (QK^T) / sqrt(dk) =\n", scaled_dot)
print("\nAttention Weights (A) =\n", A)
print("\nFinal Context Embedding (Z) =\n", Z)
```
