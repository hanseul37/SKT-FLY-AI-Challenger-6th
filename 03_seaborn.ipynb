# Seaborn

## Line Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
sns.lineplot(x = x, y = y)
sns.lineplot(x = x, y = z)
plt.title('sin and cos')
plt.show()
```

## Bar chart

```python
cat = ['A', 'B', 'C', 'D']
val = [10, 15, 8, 12]

sns.barplot(x = cat, y = val)
```

```python
cat = ['A', 'B', 'C', 'D']
val = [10, 15, 8, 12]

sns.barplot(x = cat, y = val, hue = cat)
for i, v in enumerate(val):
  plt.text(i, v - 2, str(v), fontsize = 15, color = 'white', ha='center')

plt.show()
```

```python
cat = ['A', 'B', 'C', 'D']
val = [10, 15, 8, 12]

sns.barplot(x = cat, y = val, palette = 'viridis')
for i, v in enumerate(val):
  plt.text(i, v - 2, str(v), fontsize = 15, color = 'white', ha='center')

plt.show()
```

```python
cat = ['A', 'B', 'C', 'D']
val = np.array([10, 15, 8, 12])
val = np.sort(val)

sns.barplot(x = val, y = cat)
```

## Count Plot

```python
data = ['A', 'B', 'A', 'C', 'B', 'C', 'A']

sns.countplot(x = data)
```

## Histogram

```python
data = np.random.randn(1000)

sns.histplot(data, bins = 40, kde = True)
```

## Scatter Plot

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('iris')
data
```

```python
data.columns
```

```python
sns.scatterplot(x = data['petal_length'],
                y = data['petal_width'],
                data = data,
                hue = data['species'],
                style = 'species')
```

```python
sns.pairplot(data = data,
             hue = 'species')
```

## Box Plot

```python
sns.boxplot(data = data,
            x = 'species',
            y = 'sepal_width',
            hue = 'species')
```

## Violin Plot

```python
sns.violinplot(data = data,
               x = 'species',
               y = 'sepal_width')
```
