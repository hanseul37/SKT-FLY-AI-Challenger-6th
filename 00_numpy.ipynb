```python
arr = np.array([1, 2, 3, 4, 5])
arr
```

```python
arr[0]
```

```python
arr[-1]
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
arr
```

```python
arr[0]
```

```python
arr[0][0]
```

```python
arr[0, 0]
```

```python
arr[0, ]
```

- 불리언 인덱싱(Boolean indexing)

```python
arr = np.array([1, 2, 3, 4, 5])
arr[1:3]
```

```python
arr = np.array([1, 2, 3, 4, 5])
b = [False, True, False, True, False]
arr[b]
```

```python
arr > 3
```

```python
arr[arr > 3]
```

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 > 3
```

```python
arr2[arr2 > 2]
```

```python
arr = np.arange(10)
arr[3:]
```

```python
arr[3:-1]
```

```python
arr[:]
```

```python
arr[-4:]
```
