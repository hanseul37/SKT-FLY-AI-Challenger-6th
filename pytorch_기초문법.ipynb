```python
import torchvision.transforms as transforms
```

```python
mnist_transforms = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,),(1.0,))
])
```

```python
from torchvision.datasets import MNIST
import requests
```

```python
download_root = './MNIST_DATASET'
train_dataset = MNIST(download_root, transform=mnist_transforms, train=True, download=True)
valid_dataset = MNIST(download_root, transform=mnist_transforms, train=False, download=True)
test_dataset = MNIST(download_root, transform=mnist_transforms, train=False, download=True)
```

```python
import torch
from torch.optim import SGD  # 확률적 경사 하강법(SGD) 최적화 알고리즘
from torch.nn import MSELoss  # 평균 제곱 오차 손실 함수
from torch.optim.lr_scheduler import LambdaLR  # 학습률 스케줄러

# 손실 함수 정의: 평균 제곱 오차(MSE)
criterion = MSELoss()

# 모델의 파라미터 정의 (여기서는 예시로 model.parameters() 사용)
model = torch.nn.Linear(10, 1)  # 입력 크기 10, 출력 크기 1의 선형 모델
optimizer = SGD(model.parameters(), lr=0.01, momentum=0.9)
# SGD 최적화: 학습률 0.01, 모멘텀 0.9를 사용

# 학습률 스케줄러 정의
scheduler = LambdaLR(optimizer, lr_lambda=lambda epoch: 0.95 ** epoch)
# LambdaLR: 매 에포크마다 학습률을 0.95의 거듭제곱만큼 감소

# 학습 루프
for epoch in range(1, 101):  # 총 100 에포크 동안 학습
    # 데이터 로더를 통해 배치 단위로 데이터 가져오기 (여기서는 가상 데이터)
    for x, y in [(torch.randn(32, 10), torch.randn(32, 1))]:  # 배치 크기 32
        optimizer.zero_grad()  # 이전 배치에서 계산된 기울기(gradient) 초기화
        loss = criterion(model(x), y)  # 예측 값과 실제 값의 손실 계산
        loss.backward()  # 기울기 계산
        optimizer.step()  # 파라미터 업데이트

    # 에포크마다 학습률 업데이트
    scheduler.step()
```
