```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
!wget -q https://docs.opencv.org/3.4/j.png
import os
if not os.path.exists('j.png'):
    print("j.png download failed. Please upload manually or change filename.")
else:
    print("j.png downloaded.")
```

```python
img_path = 'j.png'  # 혹은 원하는 이미지
img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)
if img is None:
    print("Image load failed.")
else:
    # kernel 정의
    kernel = np.ones((6,6), np.uint8)

    erosion = cv.erode(img, kernel, iterations=1)
    dilation = cv.dilate(img, kernel, iterations=1)
    opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
    closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

    fig, axs = plt.subplots(1,5, figsize=(18,5))
    axs[0].imshow(img, cmap='gray'); axs[0].set_title('Original'); axs[0].axis('off')
    axs[1].imshow(erosion, cmap='gray'); axs[1].set_title('Erosion'); axs[1].axis('off')
    axs[2].imshow(dilation, cmap='gray'); axs[2].set_title('Dilation'); axs[2].axis('off')
    axs[3].imshow(opening, cmap='gray'); axs[3].set_title('Opening'); axs[3].axis('off')
    axs[4].imshow(closing, cmap='gray'); axs[4].set_title('Closing'); axs[4].axis('off')
    plt.show()
```

```python
if img is not None:
  gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, kernel)
  tophat = cv.morphologyEx(img, cv.MORPH_TOPHAT, kernel)
  blackhat = cv.morphologyEx(img, cv.MORPH_BLACKHAT, kernel)

  fig, axs = plt.subplots(1,4, figsize=(16,4))
  axs[0].imshow(img, cmap='gray'); axs[0].set_title('Original'); axs[0].axis('off')
  axs[1].imshow(gradient, cmap='gray'); axs[1].set_title('Gradient'); axs[1].axis('off')
  axs[2].imshow(tophat, cmap='gray'); axs[2].set_title('Top-Hat'); axs[2].axis('off')
  axs[3].imshow(blackhat, cmap='gray'); axs[3].set_title('Black-Hat'); axs[3].axis('off')
  plt.show()
```
