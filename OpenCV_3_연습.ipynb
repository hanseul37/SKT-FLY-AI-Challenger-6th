```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
uploaded = files.upload()

filename = list(uploaded.keys())[0]

image = cv2.imread(filename)

print('원본 이미지')
cv2_imshow(image)
print('이미지 크기', image.shape)
```

```python
methods = {
    'INTER_NEAREST (최근접 이웃)': cv2.INTER_NEAREST,
    'INTER_LINEAR (양선형)': cv2.INTER_LINEAR,
    'INTER_CUBIC (상차)': cv2.INTER_CUBIC,
    'INTER_AREA (영역)': cv2.INTER_AREA,
    'INTER_LANCZOS4': cv2.INTER_LANCZOS4
}

for name, method in methods.items():
  resized = cv2.resize(
    image,
    None,
    fx=2.0,
    fy=2.0,
    interpolation=method
  )
  print(name)
  cv2_imshow(resized)
```

```python
shrink = cv2.resize(
    image,
    None,
    fx=0.8,
    fy=0.8,
    interpolation=cv2.INTER_AREA
)

cv2_imshow(shrink)
print('축소 후 이미지 크기:', shrink.shape)
```

```python
heigh, width = image.shape[:2]

m = np.float32(([1, 0, 50], [0, 1, 10]))

translated = cv2.warpAffine(
    image,
    m,
    (width, heigh)
)

cv2_imshow(translated)
```

```python
height, weight = image.shape[:2]

center = (width / 2, height / 2)

m = cv2.getRotationMatrix2D(
    center=center,
    angle=90,
    scale=0.5
    )

rotated = cv2.warpAffine(
    image,
    m,
    (width, height)
)

cv2_imshow(rotated)
```

```python
command = input()

methods = {
    'nearest': cv2.INTER_NEAREST,
    'linear': cv2.INTER_LINEAR,
    'cubic': cv2.INTER_CUBIC,
    'area': cv2.INTER_AREA,
    'lanczos4': cv2.INTER_LANCZOS4
}

resized = cv2.resize(
    image,
    None,
    fx=2.0,
    fy=2.0,
    interpolation=methods[command]
)

cv2_imshow(resized)
```

```python
fx, fy = map(float, input().split())

heigh, width = image.shape[:2]

m = np.float32(([1, 0, fx], [0, 1, fy]))

translated = cv2.warpAffine(
    image,
    m,
    (width, heigh)
)

cv2_imshow(translated)
```

```python
angle, scale = map(float, input().split())

height, weight = image.shape[:2]

center = (width / 2, height / 2)

m = cv2.getRotationMatrix2D(
    center=center,
    angle=angle,
    scale=scale
    )

rotated = cv2.warpAffine(
    image,
    m,
    (width, height)
)

cv2_imshow(rotated)
```

```python
height, width = image.shape[:2]
center = (width / 2, height / 2)

translation_matrix = np.float32([[1, 0, 50], [0, 1, 10]])

rotation_matrix = cv2.getRotationMatrix2D(center, 90, 0.5)

rotation_matrix = np.vstack([rotation_matrix, [0, 0, 1]])

composite_matrix = rotation_matrix @ translation_matrix

composite_transformed = cv2.warpAffine(image, composite_matrix[:2], (width, height))
```

```python
translated_image = cv2.warpAffine(image, translation_matrix, (width, height))
multiple_transformed = cv2.warpAffine(translated_image, rotation_matrix, (width, height))
cv2_imshow(multiple_transformed)
```
