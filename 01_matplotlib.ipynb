# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 2, 4, 6, 8])
z = x ** 2

plt.plot(x, y, label='Score')
plt.plot(x, z, label='Score2')
plt.title('This is title')
plt.xlabel('X-value')
plt.ylabel('Y-value')
plt.legend()
plt.show()
```

## Bar Chart

```python
x = ['A', 'B', 'C', 'D']
y = [10, 30, 20, 40]

plt.bar(x, y)
```

```python
plt.style.use('ggplot')
for i, v in enumerate(y):
  plt.text(i, v + 0.5, str(v), ha='center')
plt.ylim([0, 50])
```

## 다중 플롯과 서브 플롯

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
z = [0, -1, -4, -9, -16]
```

```python
plt.plot(x, y)
plt.plot(x, z)
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.subplot(2, 1, 2)
plt.plot(x, z)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.subplot(2, 2, 2)
plt.plot(x, z)
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.subplot(2, 2, 4)
plt.plot(x, z)
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(x, y)
axes[1, 1].plot(x, y)
```
